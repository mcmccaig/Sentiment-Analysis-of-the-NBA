```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
from datetime import datetime
import seaborn as sns
```

```python
raptors = pd.read_csv('../data/raptors_results.csv')
knicks = pd.read_csv('../data/knicks_results.csv')
lakers = pd.read_csv('../data/lakers_results.csv')
rockets = pd.read_csv('../data/rockets_results.csv')
warriors = pd.read_csv('../data/warriors_results.csv')
```

```python
raptors['Date'] = pd.to_datetime(raptors['Date'])
lakers['Date'] = pd.to_datetime(lakers['Date'])
knicks['Date'] = pd.to_datetime(knicks['Date'])
rockets['Date'] = pd.to_datetime(rockets['Date'])
warriors['Date'] = pd.to_datetime(warriors['Date'])
```

# Attendance

```python
#raptors
fig, ax = plt.subplots(figsize=(25,10))
sns.lineplot(x = 'Date', y = 'Attendance Ratio', data=raptors)
sns.scatterplot(x = 'Date', y = 'Attendance Ratio', data=raptors, hue='At Home', s=100, hue_order=['Home', 'Away'])
for index, row in raptors.iterrows():
    if (raptors['Attendance Ratio'][index] < 1.0):
        if(raptors['At Home'][index] == 'Away'):
            ax.text(raptors['Date'][index], raptors['Attendance Ratio'][index]-0.01, raptors['Home/Neutral'][index], va='bottom', ha='center', size='medium', weight='bold', bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=1))
    else: continue
plt.title('Raptors Attendance for the 2018-2019 Season', fontsize='20')
plt.xlabel('Date', fontsize='15')
plt.ylabel('Attendance Ratio', fontsize='15')
ax.legend(loc='lower right')
plt.grid()

fig.savefig('../Presentation/raptors_attendance.png')
```

# Polarized Sentiment Distribution

```python
def polarized_graph(team_df, team_name):
    fig, ax = plt.subplots(figsize=(25,10))
    sns.lineplot(x = 'Date', y = 'Polarized Sentiment Ratio', data=team_df, marker='o', color='blue')
    plt.axhline(0.5, color='black')
    plt.title(f'{team_name} Sentiment for the 2018-2019 Season', fontsize='20')
    plt.xlabel('Date', fontsize='15')
    plt.ylabel('Polarized Sentiment Ratio', fontsize='15')
    plt.grid()
    fig.savefig(f'../Presentation/{team_name}_sentiment.png')
```

```python
polarized_graph(raptors, 'Toronto Raptors')
```

```python
polarized_graph(knicks, 'New York Knicks')
```

```python
polarized_graph(lakers, 'Los Angeles Lakers')
```

```python
polarized_graph(rockets, 'Houston Rockets')
```

```python
polarized_graph(warriors, 'Golden State Warriors')
```

# Results Graph

```python
def results_graph(team_df, team_name):
    fig, ax = plt.subplots(figsize=(25,10))
    sns.regplot(x = 'Polarized Sentiment Ratio', y = 'Attendance Ratio', data=team_df, marker='o', color='blue')
    plt.title(f'{team_name} Attendance as a function of Sentiment', fontsize='20')
    plt.xlabel('Polarized Sentiment Ratio', fontsize='15')
    plt.ylabel('Attendance Ratio', fontsize='15')
    plt.grid()
    fig.savefig(f'../Presentation/{team_name}_results.png')
```

```python
results_graph(raptors, 'Toronto Raptors')
```

```python
results_graph(knicks, 'New York Knicks')
```

```python
results_graph(lakers, 'Los Angeles Lakers')
```

```python
results_graph(rockets, 'Houston Rockets')
```

```python
results_graph(warriors, 'Golden State Warriors')
```
