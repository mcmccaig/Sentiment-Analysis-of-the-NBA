```python
import pandas as pd
import numpy as np
```

#### Get Canucks Attendance

```python
attendance = pd.read_csv('../data/canucks/canucks_201819_schedule.csv')
```

```python
attendance
```

#### Canucks Capacity

```python
capacity = pd.read_csv('../data/canucks/arena_capacity.csv')
```

```python
capacity.head()
```

```python
capacity_dict = dict(zip(capacity['Home Team'], capacity['Capacity(Seating capacity only)']))
capacity_dict
```

```python
attendance['Unnamed: 2'] = attendance['Unnamed: 2'].fillna('Home')
```

```python
attendance
```

```python
attendance['Capacity'] = attendance['Opponent']
```

```python
home_mask = attendance['Unnamed: 2'] == 'Home'
away_mask = attendance['Unnamed: 2'] == '@'
```

```python
attendance['Capacity'][home_mask] = capacity_dict['Vancouver Canucks']
```

```python
attendance['Capacity'][away_mask] = attendance['Capacity'][away_mask].map(capacity_dict)
```

```python
attendance
```

```python
attendance['Capacity'] = attendance['Capacity'].str.replace(',', '')
```

```python
attendance['Capacity'] = pd.to_numeric(attendance['Capacity'])
```

```python
type(attendance['Capacity'][1])
```

```python
attendance['Attendance Ratio'] = attendance['Att.']/attendance['Capacity']
```

```python
attendance
```

```python
mask = attendance['Attendance Ratio'] > 1
attendance['Attendance Ratio'] = attendance['Attendance Ratio'].where(~mask, 1)
```

```python
attendance
```

```python
attendance.to_csv('../data/canucks/canucks_201819_attendance.csv')
```
