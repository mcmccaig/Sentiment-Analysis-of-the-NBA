```python
import numpy as np
import pandas as pd
import glob

pd.set_option('max_colwidth', 1000)
pd.set_option('max_rows', 1000)
```

```python
#create dataframe with all csvs in this folder concatenated
canucks = pd.concat([pd.read_csv(f) for f in glob.glob('../data/canucks/tweets/*.csv')], ignore_index = True)
```

```python
canucks
```

```python
#check for NaN
canucks.isna().sum()
```

```python
canucks.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
canucks['Date'] = pd.to_datetime(canucks['Date'])
PST = pd.Timedelta('8 hours')
canucks['Date'] = canucks['Date'] - PST
```

```python
#remove hyperlinks
canucks['Tweet'] = canucks['Tweet'].str.replace('\w+:\/\/\S+', '', case=False)
```

```python
#drop duplicate tweets
canucks = canucks.drop_duplicates('Tweet')
```

```python
canucks.reset_index(drop=True, inplace=True)
```

```python
canucks
```

```python
canucks.to_csv('../data/canucks/tweets/canucks_tweets_cleaned.csv')
```
