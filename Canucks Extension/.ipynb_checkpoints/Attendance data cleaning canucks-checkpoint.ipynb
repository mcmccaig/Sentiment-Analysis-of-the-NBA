# Getting total season schedule (all teams) with attendance numbers

```python
df1 = pd.read_csv('../data/attendance/2018-2019/october_2018_nba_attendance.csv')
df2 = pd.read_csv('../data/attendance/2018-2019/november_2018_nba_attendance.csv')
df3 = pd.read_csv('../data/attendance/2018-2019/december_2018_nba_attendance.csv')
df4 = pd.read_csv('../data/attendance/2018-2019/january_2019_nba_attendance.csv')
df5 = pd.read_csv('../data/attendance/2018-2019/february_2019_nba_attendance.csv')
df6 = pd.read_csv('../data/attendance/2018-2019/march_2019_nba_attendance.csv')
df7 = pd.read_csv('../data/attendance/2018-2019/april_2019_nba_attendance.csv')

df_list = [df1,df2,df3,df4,df5,df6,df7]
```

```python
season_2018 = pd.DataFrame()
```

```python
#combine all seasons into one df
for i in df_list:
    season_2018 = season_2018.append(i, ignore_index=True)
```

```python
season_2018
```

```python
# remove playoff games
season_2018.iloc[1230:, :]
season_2018 = season_2018.drop(season_2018.index[1230:])
```

```python
season_2018.head()
```

```python
capacity = pd.read_csv('../data/attendance/arena_capacity.csv')
```

```python
capacity.head()
```

```python
capacity_dict = dict(zip(capacity['Team(s)'], capacity['Capacity']))
capacity_dict
```

```python
# add arena capacity to season 2018 df
season_2018['Capacity']= season_2018['Home/Neutral'].map(capacity_dict)
season_2018
```

```python
season_2018.to_csv('../data/attendance/2018-2019/2018_2019_attendance.csv')
```

# team specific game attendance

```python
raptors_att = pd.DataFrame()
```

```python
def attendance_combiner(season_att, team_string, team_att):
    away_mask = season_att['Visitor/Neutral'] == team_string
    home_mask = season_att['Home/Neutral'] == team_string
    team_att = team_att.append(season_att[home_mask])
    team_att = team_att.append(season_att[away_mask])
    team_att['Date'] = pd.to_datetime(team_att['Date'])
    team_att['Attend.'] = pd.to_numeric(team_att['Attend.'])
    team_att['Capacity'] = pd.to_numeric(team_att['Capacity'])
    team_att['Attendance Ratio'] = (team_att['Attend.'] / team_att['Capacity'])
    return(team_att)
```

```python
#isolate raptors season information from season 2018 df
raptors_att = attendance_combiner(season_2018, 'Toronto Raptors', raptors_att)
raptors_att = raptors_att.sort_values(by='Date')
raptors_att = raptors_att.reset_index()
raptors_att
```

```python
one_mask = raptors_att['Attendance Ratio'] > 1
```

```python
raptors_att['Attendance Ratio'] = raptors_att['Attendance Ratio'].where(~one_mask, 1)
```

```python
raptors_att
```

```python
raptors_att.to_csv('../data/attendance/teams/raptors_2018_2019_attendance.csv')
```
