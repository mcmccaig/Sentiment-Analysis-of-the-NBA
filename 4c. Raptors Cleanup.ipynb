# Raptors Tweets Cleaning

```python
raptors = pd.read_csv('../data/tweets/raptors/tweets v3/raptors_players_concat_tweets.csv')
```

```python
#check for NaN
raptors.isna().sum()
```

```python
#order by date
raptors['Date'] = pd.to_datetime(raptors['Date'])
EST = pd.Timedelta('5 hours')
raptors['Date'] = raptors['Date'] - EST
raptors['Date'] = raptors['Date'].sort_values()
```

```python
#drop duplicate tweets
raptors = raptors.drop_duplicates('Tweet')
```

```python
#drop unnamed columns and reset index
raptors.drop(['Unnamed: 0'], axis=1, inplace=True)
raptors.drop(['Unnamed: 0.1'], axis=1, inplace=True)
raptors.reset_index(drop=True, inplace=True)
```
